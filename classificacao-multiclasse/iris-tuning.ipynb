```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import np_utils
from keras.optimizers import Adam
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV, cross_val_score

np.random.seed(42)
X = pd.read_csv('iris.csv').iloc[:, 0:4].values
y = pd.read_csv('iris.csv').iloc[:, 4].values

labelEncoder = LabelEncoder()
y = labelEncoder.fit_transform(y)
y_dummy = np_utils.to_categorical(y)
```

```python
def criar_rede(units_first, lr, clipvalue, decay, clipnorm):
    model = Sequential()
    
    model.add(Dense(units=units_first, activation='relu', input_dim=4))
    model.add(Dense(units=3, activation='softmax'))
    
    optimizer = Adam(lr=lr, decay=decay, clipvalue=clipvalue, clipnorm=clipnorm)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
model = KerasClassifier(build_fn=criar_rede, epochs=100, batch_size=10)
parametros = {"lr": np.linspace(0.0001, 0.1, num=3),
              "decay": np.linspace(0.0001, 0.01, num=2),
              "clipnorm": np.linspace(0.1, 1, num=2),
              "units_first": range(4, 15),
              "clipvalue": np.linspace(-0.5, 0.5, num=2)}

grid = GridSearchCV(estimator=model, param_grid=parametros)
grid = grid.fit(X,y)
```

```python
print(f'{grid.best_estimator_}')
```
