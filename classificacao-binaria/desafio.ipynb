```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
X = pd.read_csv('entradas-breast.csv')
y = pd.read_csv('saidas-breast.csv')
```

```python
def criar_rede():
    model = Sequential()

    model.add(Dense(units=8,
            activation='relu',
            kernel_initializer='random_uniform',
            input_dim=30))

    model.add(Dropout(0.2))
    model.add(Dense(units=32,
              activation='relu',
              kernel_initializer='random_uniform'))
    
    model.add(Dropout(0.2))
    model.add(Dense(units=32,
              activation='relu',
              kernel_initializer='random_uniform'))
    
    model.add(Dropout(0.2))
    model.add(Dense(units=8,
            activation='relu',
            kernel_initializer='random_uniform'))

    model.add(Dropout(0.2))
    model.add(Dense(units=1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])

    return model
```

```python
model = KerasClassifier(build_fn = criar_rede,
                        batch_size= 10,
                        epochs = 500)
score = cross_val_score(estimator=model, X=X, y=y, cv=10, scoring= 'accuracy')
```

```python
print(f'desvio padro: {score.mean()}')
print(f'desvio padro: {score.std()}')
```
