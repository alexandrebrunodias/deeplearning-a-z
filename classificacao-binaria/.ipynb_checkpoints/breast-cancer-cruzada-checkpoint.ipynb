```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
X = pd.read_csv('entradas-breast.csv')
y = pd.read_csv('saidas-breast.csv')
```

```python
def criarRede():
    model = Sequential()

    # PRIMEIRA CAMADA OCULTA
    model.add(Dense(units=16,
                    activation='relu', 
                    kernel_initializer='random_uniform',
                    input_dim=30))
    # CAMADA DE DROPOUT
    model.add(Dropout(0.2))
    
    # SEGUNDA CAMADA OCULTA
    model.add(Dense(units=16, 
                    activation='relu', 
                    kernel_initializer='random_uniform'))
    
    # CAMADA DE DROPOUT
    model.add(Dropout(0.2))
    
    model.add(Dense(units=1, activation='sigmoid'))
    
    # CONFIRGURANDO OTIMIZADOR
    otimizador = keras.optimizers.Adam(lr=0.001, 
                                       decay=0.0001,
                                       clipvalue=0.5)
    model.compile(optimizer=otimizador,
                  loss='binary_crossentropy',
                  metrics=['binary_accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=criarRede,
                        epochs=100,
                        batch_size=10)
result = cross_val_score(estimator=model,
                        X=X, 
                        y=y,
                        cv=10,
                        scoring='accuracy')
```
