```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# PREPROCESSAMENTO

df = pd.read_csv('games.csv')
df = df.drop('Other_Sales', axis = 1)
df = df.drop('Global_Sales', axis = 1)
df = df.drop('Developer', axis = 1)

df = df.dropna(axis = 0)
df = df.loc[df['NA_Sales'] > 1]
df = df.loc[df['EU_Sales'] > 1]

df['Name'].value_counts()
nome_jogos = df.Name
df = df.drop('Name', axis = 1)

X = df.iloc[:, [0,1,2,3,7,8,9,10,11]].values

labelEncoder = LabelEncoder()
X[: ,0] = labelEncoder.fit_transform(X[: ,0])
X[: ,2] = labelEncoder.fit_transform(X[: ,2])
X[: ,3] = labelEncoder.fit_transform(X[: ,3])
X[: ,8] = labelEncoder.fit_transform(X[: ,8])

one_hot_encoder = OneHotEncoder(categorical_features=[0, 2, 3 ,8])
X = one_hot_encoder.fit_transform(X).toarray()

y1 = df.iloc[:, 4].values
y2 = df.iloc[:, 5].values
y3 = df.iloc[:, 6].values
```

```python
inputs = tf.keras.Input(shape=(61,))
layer1 = tf.keras.layers.Dense(units=32, activation=tf.nn.sigmoid)(inputs)
layer2 = tf.keras.layers.Dense(units=32, activation=tf.nn.sigmoid)(layer1)

output_layer1 = tf.keras.layers.Dense(units=1, activation='linear')(layer2)
output_layer2 = tf.keras.layers.Dense(units=1, activation='linear')(layer2)
output_layer3 = tf.keras.layers.Dense(units=1, activation='linear')(layer2)

model = tf.keras.Model(inputs=inputs, outputs=(output_layer1, output_layer2, output_layer3))

model.compile(optimizer=tf.optimizers.Adam(), loss=tf.metrics.mse)

model.fit(X, [y1, y2, y3], epochs=1000, batch_size=10)
```

```python
from sklearn.metrics import r2_score

y1_pred, y2_pred, y3_pred = model.predict(X)

s1 = r2_score(y1, y1_pred)
s2 = r2_score(y2, y2_pred)
s3 = r2_score(y3, y3_pred)

print(f's1: {s1} \ns2: {s2} \ns3: {s3} \nMédia: {(s1 + s2 + s3)/3}')
```
