```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# PREPROCESSAMENTO
df = pd.read_csv('games.csv')

df = df.drop('Other_Sales', axis = 1)
df = df.drop('NA_Sales', axis = 1)
df = df.drop('EU_Sales', axis = 1)
df = df.drop('JP_Sales', axis =1)
df = df.drop('Developer', axis = 1)

df = df.dropna(axis = 0)
df = df.loc[df['Global_Sales'] > 1]

df['Name'].value_counts()
nome_jogos = df.Name
df = df.drop('Name', axis = 1)

X = df.iloc[:, [0,1,2,3,5,6,7,8,9]].values

labelEncoder = LabelEncoder()
X[: ,0] = labelEncoder.fit_transform(X[: ,0])
X[: ,2] = labelEncoder.fit_transform(X[: ,2])
X[: ,3] = labelEncoder.fit_transform(X[: ,3])
X[: ,8] = labelEncoder.fit_transform(X[: ,8])

one_hot_encoder = OneHotEncoder(categorical_features=[0, 2, 3 ,8])
X = one_hot_encoder.fit_transform(X).toarray()

y = df.iloc[:, 4].values
```

```python
inputs = tf.keras.Input(shape=(99,))
layer1 = tf.keras.layers.Dense(units=8, activation=tf.nn.relu)(inputs)
layer2 = tf.keras.layers.Dense(units=8, activation=tf.nn.relu)(layer1)

output = tf.keras.layers.Dense(units=1, activation='linear')(layer2)

model = tf.keras.Model(inputs=inputs, outputs=(output))

model.compile(optimizer=tf.optimizers.Adam(), loss=tf.metrics.mse)

model.fit(X, y, epochs=5000, batch_size=10)
```

```python
# Teste simples
from sklearn.metrics import r2_score

y_pred = model.predict(X)

print(f'Score: {r2_score(y, y_pred)}')
```
